```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import  mean_squared_error
```

```python
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv")
```

```python
column_names = {'T1':'Kitchen_T', 'RH_1': 'Kitchen_H', 'T2':  'Livingroom_T', 'RH_2': 'Livingroom_H', 'T3': 'Laundryroom_T',
                'RH_3': 'Laundryroom_H', 'T4': 'Officeroom_T', 'RH_4': 'Officeroom_H', 'T5': 'Bathroom_T', 'RH_5': 'Bathroom_H',
                'T6':'Outsidenorth_T', 'RH_6': 'Outsidenorth_H', 'T7': 'Ironingroom_T', 'RH_7': 'Ironingroom_H', 'T8': 'Teenagerroom_T',
                'RH_8': 'Teenagerroom_H', 'T9': 'Parentroom_T', 'RH_9': 'Parentroom_H', 'T_out': 'Outsidechievres_T', 'Press_mm_hg':'Pressure',
                'RH_out': 'Outsidechievres_H'}
```

```python
data = data.rename(columns=column_names)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
```

```python
#R^2 score
lm.score(X, Y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data = data.drop(columns=['date', 'lights'])
normalised_data = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
features_data = normalised_data.drop(columns=['Appliances'])
target = normalised_data['Appliances']


#Now, we split our dataset into the training and testing dataset. Recall that we had earlier segmented the features and target variables.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_data, target, test_size=0.3, random_state=42)

lm =linear_model.LinearRegression()
#fit the model to the training dataset
lm.fit(x_train, y_train)
#obtain predictions
predicted_values = lm.predict(x_test)


#Mean Absolute Error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)	    
```

```python
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 2)
```
